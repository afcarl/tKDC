```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("home_dataset/HT_Sensor_Dataset.dat", delim_whitespace=True)
```

```python
df = df.reset_index()
np.random.seed(0)
shuffled = df.reindex(np.random.permutation(df.index))
shuffled = shuffled.drop(["index"], axis=1)
```

```python
shuffled.to_csv("shuffled_homesensor.csv", index=False)
```
